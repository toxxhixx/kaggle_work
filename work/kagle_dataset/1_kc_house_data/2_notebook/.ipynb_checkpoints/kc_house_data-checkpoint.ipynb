# House Sales in King County, USA
https://www.kaggle.com/harlfoxem/housesalesprediction/data

```python
import pandas as pd
from IPython.display import display
from dateutil.parser import parse
import matplotlib.pyplot as plt
```

```python
df_data = pd.read_csv("kc_house_data.csv")
print(df_data.columns)
display(df_data.head())
display(df_data.tail())
```

```python
# coutn missing
pd.DataFrame(df_data.isnull().sum(), columns=["num of missing"])
```

```python
# date列の変換
df_data["date"] = [ parse(i[:-7]).date() for i in df_data["date"]]
display(df_data.head())
```

```python
# describe
df_data.describe()
```

```python
df_data2 = df_data.drop(["id","date","zipcode"], axis=1)
df1 = df_data2.iloc[:,:9]
display(df1.head())
df2 = df_data2.iloc[:,[0]+list(range(9,18))]
display(df2.head())
```

```python
# 散布図行列
pd.plotting.scatter_matrix(df1,figsize=(13,13))
plt.show()
pd.plotting.scatter_matrix(df2,figsize=(13,13))
plt.show()
```

```python
# histogram
for col in df_data2.columns:
    print(col)
    df_data2[col].hist()
    plt.xlabel(col)
    plt.ylabel("num")
    plt.show()
```
